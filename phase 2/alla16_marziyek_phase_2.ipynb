**2.3 Text Data Cleaning**
- Lowercasing words
- Removing stopwords
- Removing Punctuations
- Removing other characters that are not relevant
- Removing extra spaces

```python
def clean_text_data(sentence, stop_words):
    sentence = sentence.lower()  # converting the text into lower case to standardize the text
    sentence = sentence.split()
    text = ""
    for word in sentence:  # removing stopwords from the sentence
        if word in stop_words:
            continue
        else:
            text = text + word + " "
    text = text.strip()
    punctuations = string.punctuation
    text = ''.join([punc for punc in text if punc not in punctuations])  # removing punctuations from the sentence
     # there are different symbols like â¥, Ã§ÂºÂ½Ã§ÂºÂ¦Ã¤Â¹â€¹Ã¥Â®Â¶ in the data, hence to remove them, we are just considering the text that is in A-Z, a-z or digits
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) 
    text = ' '.join(text.split())
    return text
```

```python
stop_words = stopwords.words('english')
```

```python
df['name'] = df['name'].apply(lambda x: clean_text_data(str(x),stop_words))  # cleaning text data in name column
df['host_name'] = df['host_name'].apply(lambda x: str(x).lower())  # lowercasing all names in host_name column
df["room_type"] = df["room_type"].str.lower().str.replace('/',' or ')  # Normalizing text data in room type column (to replace special character '/' with or)
df['neighbourhood_group'] = df['neighbourhood_group'].apply(lambda x: clean_text_data(str(x),stop_words))  # cleaning - lowercasing, punctuations
df['neighbourhood'] = df['neighbourhood'].str.lower()  # lowercasing all neighbourhood data
```

```python
# Handling Outliers
def remove_outliers_using_iqr(df, col_name):
    count = 0
    q1 = df[col_name].quantile(0.25)
    q3 = df[col_name].quantile(0.75)
    iqr = q3-q1
    lowerbound = q1 - 1.5 * iqr
    upperbound = q3 + 1.5* iqr
    outliers = (df[col_name] < lowerbound) | (df[col_name] > upperbound)
    df = df[~outliers]
    return df
```

```python
for col in df.select_dtypes(include=['number']):
    min = df[col].min()
    max = df[col].max()
    print(f'col: {col}, min: {min}, max: {max}')
```

```python
numerical_columns = ['price', 'minimum_nights','number_of_reviews', 'availability_365']  # we are removing outliers from all numerical column so that our model should not be 
for col in numerical_columns:
    df = remove_outliers_using_iqr(df, col)
```

```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import linear_model
import matplotlib as mpl
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn import naive_bayes
from sklearn import tree
import xgboost as xgb
import seaborn as sns
from sklearn.neighbors import KNeighborsRegressor
import warnings
warnings.filterwarnings('ignore')
mpl.rc('font',family='Times New Roman')
```

```python
def calculate_metrics(y_true, y_pred):
    mpe = abs(np.mean((y_true - y_pred) / y_true) * 100)
    mse = mean_squared_error(y_true , y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return mpe,mse,r2,rmse
```

```python
df = pd.read_csv('./not_normal_data.csv', index_col=False) # reading data that is not normalized
```

```python
corr_matrix = df[['neighbourhood_group','latitude','longitude','room_type', 'minimum_nights', 'number_of_reviews','reviews_per_month','calculated_host_listings_count', 'availability_365','price']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## Ordinary least squares (Linear Regression)

```python
numerical_columns = ['neighbourhood_group','latitude','longitude','room_type',
                     'minimum_nights', 'number_of_reviews','reviews_per_month',
                     'calculated_host_listings_count', 'availability_365']

X = df[numerical_columns].to_numpy()
y = df['price'].to_numpy()
pre  = sklearn.preprocessing.MinMaxScaler(feature_range=(0, 1))
X = pre.fit_transform(X)

# splitting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# training model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)

# Plotting actual vs predicted values
plt.figure(figsize=(6, 6),layout='tight')
plt.scatter(y_test, y_pred, color='b',s=15)
plt.plot(y_test, y_test, color='black')
plt.title("Actual vs. Predicted Prices, method: LinearRegression")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.xlim(-30,350)
plt.ylim(-30,350)
plt.text(0,335,f"RÂ² Score:{round(r_squared,4)}")
plt.savefig('./lr_pred.png')

# calculating the performance metrics
mpe,mse,r2,rmse = calculate_metrics(y_test, y_pred)
print('Metrics: ')
print(f"Mean Percentage Error:{round(mpe,4)}")
print(f"Root Mean Squared Error:{round(rmse,4)}")
print(f"Mean Squared Error:{round(mse,4)}")
print(f"RÂ² Score:{round(r2,4)}")
```

## Naive Bayes classifier

```python
X = df[numerical_columns].to_numpy()
y = df['price'].to_numpy()
pre  = sklearn.preprocessing.MinMaxScaler(feature_range=(0, 1))
X = pre.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# training the model
model = naive_bayes.MultinomialNB(fit_prior=False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)

# plotting actual vs predicted values
plt.figure(figsize=(6, 6),layout='tight')
plt.scatter(y_test, y_pred, color='b',s=15)
plt.plot(y_test, y_test, color='black')
plt.title("Actual vs. Predicted Prices, method: NaiveBayes")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.xlim(-30,350)
plt.ylim(-30,350)
plt.text(0,335,f"RÂ² Score:{round(r_squared,4)}")
plt.savefig('./naive_bayes.png')

# calculating the performance metrics
mpe,mse,r2,rmse = calculate_metrics(y_test, y_pred)
print('Metrics: ')
print(f"Mean Percentage Error:{round(mpe,4)}")
print(f"Root Mean Squared Error:{round(rmse,4)}")
print(f"Mean Squared Error:{round(mse,4)}")
print(f"RÂ² Score:{round(r2,4)}")
```

- Decision Trees are non-parametric models, which means they don't depend on the scale of the features. Therefore, you don't necessarily need to normalize the features.

```python
from sklearn.neural_network import MLPRegressor

df = pd.read_csv('./not_normal_data.csv', index_col=False)

numerical_columns = ['neighbourhood_group','latitude','longitude','room_type',
                     'minimum_nights', 'number_of_reviews','reviews_per_month',
                     'calculated_host_listings_count', 'availability_365']

X = df[numerical_columns].to_numpy()
y = df['price'].to_numpy()
pre  = sklearn.preprocessing.MinMaxScaler(feature_range=(0, 1))
X = pre.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = MLPRegressor(hidden_layer_sizes=(100,), batch_size=16,activation='relu', 
                     solver='adam',learning_rate_init=0.001,learning_rate='adaptive',
                     max_iter=500,early_stopping=True)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)

plt.figure(figsize=(6, 6),layout='tight')
plt.scatter(y_test, y_pred, color='b',s=15)
plt.plot(y_test, y_test, color='black')
plt.title("Actual vs. Predicted Prices, method: NeuralNetwork")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.xlim(-30,350)
plt.ylim(-30,350)
plt.text(0,335,f"RÂ² Score:{round(r_squared,4)}")
plt.savefig('./NeuralNetwork.png')


mpe,mse,r2,rmse = calculate_metrics(y_test, y_pred)
print('Metrics: ')
print(f"Mean Percentage Error:{round(mpe,4)}")
print(f"Root Mean Squared Error:{round(rmse,4)}")
print(f"Mean Squared Error:{round(mse,4)}")
print(f"RÂ² Score:{round(r2,4)}")
```
